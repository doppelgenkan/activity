```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('memdata_TSND121-15082369_2019-07-09_101823.965.csv',
                 names=['item', 'time', 'a_x', 'a_y', 'a_z', 'u_x', 'u_y','u_z'])
```

```python
a_x = np.array(df['a_x'])
```

```python
F_a_x = np.fft.fft(a_x[:2**12])/len(F_a_x)
print(len(F_a_x))
```

```python
plt.plot(np.arange(0, len(F_a_x))*1000/len(F_a_x), np.abs(F_a_x))
```

```python
hz = np.arange(len(F_a_x))*1000/len(F_a_x)
```

```python
plt.plot(hz[1:500], np.abs(F_a_x[1:500]))
```

```python
F_a_y = np.array(df['a_y'])[:2**12]
F_a_y = np.fft.fft(a_y)/len(F_a_y)
```

```python
plt.plot(hz[50:200], np.abs(F_a_x[50:200])**2)
```

```python
F_a_z = np.array(df['a_z'])
F_a_z = np.fft.fft(a_y)/len(F_a_z)
```

```python
plt.plot(hz[1:3000], np.abs(F_a_z[1:3000])**2)
```
