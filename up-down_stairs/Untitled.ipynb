```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('memdata_TSND121-15082369_2019-07-09_101823.965.csv',
                 names=['item', 'time', 'a_x', 'a_y', 'a_z', 'u_x', 'u_y','u_z'])
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df['a_x'])
plt.plot(df['a_y'])
plt.plot(df['a_z'])
plt.plot(np.sqrt(np.array(df['a_x'])**2+np.array(df['a_y'])**2+np.array(df['a_z'])**2))
```

```python
a_x = np.array(df['a_x'])[:2**15]
```

```python
plt.plot(a_x)
```

```python
F_a_x = np.fft.fft(a_x[:2**15])/len(a_x)
print(len(F_a_x))
```

```python
plt.plot(np.arange(0, len(F_a_x))*1000/len(F_a_x), np.abs(F_a_x))
plt.xlabel('frequency [Hz]')
plt.ylabel('square root of power')
```

```python
hz = np.arange(len(F_a_x))*1000/len(F_a_x)
```

```python
plt.plot(hz[1:500], np.abs(F_a_x[1:500])**2)
plt.xlabel('frequency [Hz]')
plt.ylabel('power')
```

```python
a_y = np.array(df['a_y'])[:2**15]
F_a_y = np.fft.fft(a_y)/len(a_y)
```

```python
plt.plot(hz[100:1000], np.abs(F_a_y[100:1000])**2)
plt.xlabel('frequency [Hz]')
plt.ylabel('power')
```

```python
a_z = np.array(df['a_z'])[:2**15]
F_a_z = np.fft.fft(a_z)/len(F_a_z)
```

```python
plt.plot(hz[1:], np.abs(F_a_z[1:])**2)
```

```python
plt.plot(hz[100:1000], np.abs(F_a_z[100:1000])**2)
plt.xlabel('frequency [Hz]')
plt.ylabel('power')
plt.title('$F_{a_z}$')
```

```python
plt.plot(hz[100:1000], np.abs(F_a_y[100:1000])**2)
plt.xlabel('frequency [Hz]')
plt.ylabel('power')
plt.title('$F_{a_y}$')
```

```python
plt.plot(hz[100:1000], np.abs(F_a_x[100:1000])**2)
plt.xlabel('frequency [Hz]')
plt.ylabel('power')
plt.title('$F_{a_x}$')
```
