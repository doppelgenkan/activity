```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lpf
from integral import integral
```

```python
df = pd.read_csv('memdata_TSND121-13121423_2019-05-20_142057.569.csv', header=None,
                 names=['item', 't', 'acc_x', 'acc_y', 'acc_z', 'omega_x', 'omega_y', 'omega_z'])
```

```python
df.head()
```

```python
z = np.array(df['omega_z'])/100
```

```python
zeta = lpf.lpfilter(z, 20)
```

```python
plt.figure(figsize=(40,5))
plt.plot(zeta[:60000])
```

```python
theta = integral(zeta, 0, 0, 60000)
```

```python
plt.figure(figsize=(40,5), dpi=300)
plt.plot(theta)
plt.xlabel('time [msec]')
plt.ylabel('anle[deg]')
```

```python
plt.figure(figsize=(10,10), dpi=300)
plt.plot(theta[0:50000], zeta[0:50000])
plt.title('phase space')
plt.xlabel('angle[deg]')
plt.ylabel('anglar velocity[deg/sec]')
```
